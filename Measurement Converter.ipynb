## Measurement Converter 

This notebook demonstrates a simple command-line unit converter for length measurements.

### Common Length Conversion Factors

Imperial to Imperial:
- 1 foot = 12 inches
- 1 yard = 3 feet
- 1 yard = 36 inches

Metric to Metric:
- 1 centimeter (cm) = 10 millimeters (mm)
- 1 meter (m) = 100 centimeters (cm)
- 1 meter (m) = 1000 millimeters (mm)

Imperial to Metric (approximate, for quick reference):
- 1 inch ≈ 2.5400 centimeters (cm)
- 1 inch ≈ 25.4000 millimeters (mm)
- 1 foot ≈ 30.4800 centimeters (cm)
- 1 foot ≈ 0.3048 meters (m)
- 1 yard ≈ 0.9144 meters (m)

Metric to Imperial (approximate, for quick reference):
- 1 centimeter (cm) ≈ 0.3937 inches
- 1 meter (m) ≈ 39.3701 inches
- 1 meter (m) ≈ 3.2808 feet
- 1 meter (m) ≈ 1.0936 yards

*Note: The code uses precise conversion factors, but these approximate values are provided for easy reference.*

```python
# User Input
convert_from = input ("Enter Starting Unit of Measurement(Inches, Feet, Yards, Millimeters, Centimeters, Meters): ")
convert_to = input ("Enter Unit of Measurement to Convert to(Inches, Feet, Yards, Millimeters, Centimeters, Meters): ")
```

```python
# Conditional Logic
if convert_from.lower() in ["in", "inches", "inch"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Inches: "))
        if convert_to.lower() in ["ft", "feet", "foot"]:
            print(f"{number_of_units} Inches is equal to {number_of_units / 12:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["yd", "yds", "yard", "yards"]:
            print(f"{number_of_units} Inches is equal to {number_of_units / 36:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["mm", "millimeters", "millimeter"]:
            print(f"{number_of_units} Inches is equal to {number_of_units * 25.4:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["cm", "centimeters", "centimeter"]:
            print(f"{number_of_units} Inches is equal to {number_of_units * 2.54:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["m", "meters", "meter"]:
            print(f"{number_of_units} Inches is equal to {number_of_units * 0.0254:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Inches conversion. Please enter Feet, Yards, Millimeters, Centimeters, or Meters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

elif convert_from.lower() in ["ft", "feet", "foot"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Feet: "))
        if convert_to.lower() in ["in", "inches", "inch"]:
            print(f"{number_of_units} Feet is equal to {number_of_units * 12:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["yd", "yds", "yard", "yards"]:
            print(f"{number_of_units} Feet is equal to {number_of_units / 3:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["mm", "millimeters", "millimeter"]:
            print(f"{number_of_units} Feet is equal to {number_of_units * 304.8:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["cm", "centimeters", "centimeter"]:
            print(f"{number_of_units} Feet is equal to {number_of_units * 30.48:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["m", "meters", "meter"]:
            print(f"{number_of_units} Feet is equal to {number_of_units * 0.3048:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Feet conversion. Please enter Inches, Yards, Millimeters, Centimeters, or Meters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

elif convert_from.lower() in ["yd", "yds", "yard", "yards"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Yards: "))
        if convert_to.lower() in ["in", "inches", "inch"]:
            print(f"{number_of_units} Yards is equal to {number_of_units * 36:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["ft", "feet", "foot"]:
            print(f"{number_of_units} Yards is equal to {number_of_units * 3:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["mm", "millimeters", "millimeter"]:
            print(f"{number_of_units} Yards is equal to {number_of_units * 914.4:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["cm", "centimeters", "centimeter"]:
            print(f"{number_of_units} Yards is equal to {number_of_units * 91.44:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["m", "meters", "meter"]:
            print(f"{number_of_units} Yards is equal to {number_of_units * 0.9144:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Yards conversion. Please enter Inches, Feet, Millimeters, Centimeters, or Meters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

elif convert_from.lower() in ["mm", "millimeters", "millimeter"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Millimeters: "))
        if convert_to.lower() in ["in", "inches", "inch"]:
            print(f"{number_of_units} Millimeters is equal to {number_of_units / 25.4:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["ft", "feet", "foot"]:
            print(f"{number_of_units} Millimeters is equal to {number_of_units / 304.8:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["yd", "yds", "yard", "yards"]:
            print(f"{number_of_units} Millimeters is equal to {number_of_units / 914.4:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["cm", "centimeters", "centimeter"]:
            print(f"{number_of_units} Millimeters is equal to {number_of_units / 10:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["m", "meters", "meter"]:
            print(f"{number_of_units} Millimeters is equal to {number_of_units / 1000:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Millimeters conversion. Please enter Inches, Feet, Yards, Centimeters, or Meters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

elif convert_from.lower() in ["cm", "centimeters", "centimeter"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Centimeters: "))
        if convert_to.lower() in ["in", "inches", "inch"]:
            print(f"{number_of_units} Centimeters is equal to {number_of_units / 2.54:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["ft", "feet", "foot"]:
            print(f"{number_of_units} Centimeters is equal to {number_of_units / 30.48:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["yd", "yds", "yard", "yards"]:
            print(f"{number_of_units} Centimeters is equal to {number_of_units / 91.44:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["mm", "millimeters", "millimeter"]:
            print(f"{number_of_units} Centimeters is equal to {number_of_units * 10:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["m", "meters", "meter"]:
            print(f"{number_of_units} Centimeters is equal to {number_of_units / 100:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Centimeters conversion. Please enter Inches, Feet, Yards, Millimeters, or Meters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

elif convert_from.lower() in ["m", "meters", "meter"]:
    try:
        number_of_units = float(input("Enter Starting Measurement in Meters: "))
        if convert_to.lower() in ["in", "inches", "inch"]:
            print(f"{number_of_units} Meters is equal to {number_of_units / 0.0254:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["ft", "feet", "foot"]:
            print(f"{number_of_units} Meters is equal to {number_of_units / 0.3048:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["yd", "yds", "yard", "yards"]:
            print(f"{number_of_units} Meters is equal to {number_of_units / 0.9144:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["mm", "millimeters", "millimeter"]:
            print(f"{number_of_units} Meters is equal to {number_of_units * 1000:.2f} {convert_to.title()}")
        elif convert_to.lower() in ["cm", "centimeters", "centimeter"]:
            print(f"{number_of_units} Meters is equal to {number_of_units * 100:.2f} {convert_to.title()}")
        else:
            print("Invalid target unit for Meters conversion. Please enter Inches, Feet, Yards, Millimeters, or Centimeters.")
    except ValueError:
        print("Invalid measurement entered. Please enter a numeric value.")

else:
    print("Invalid starting unit. Please enter Inches, Feet, Yards, Millimeters, Centimeters, or Meters.")
```
